```python
def ts(Xi):
    """
    Усредняем побочные диоганали элементарной матрицы 
    и переводим в временной ряд
    """
    Xrev = Xi[::-1]
    return np.array([Xrev.diagonal(i).mean() for i in range(-Xi.shape[0]+1, Xi.shape[1])])

# plt.plot(t,xToTs(X[8]))
```

```python
def ssa(F, L):
    N = len(F) 
    K = N - L + 1 # Количество столбцов в траекторной матрице
    X = np.column_stack([F[i:i+L] for i in range(0,K)])
    d = np.linalg.matrix_rank(X) 
    print("d",d)
    # Сингулярное разложение
    U, Sigma, V = np.linalg.svd(X)
    V = V.T 
    X = np.array([Sigma[i] * np.outer(U[:,i], V[:,i]) for i in range(0, L)])
    return X ,sigma
    # XElem.shape
    
```

```python
def getComponents(X, p):
    d = X.shape[0]
    components = []
    for i in range(d):
        for j in range(i+1, d):
            stat = scipy.stats.pearsonr(ts(X[i]), ts(X[j]))[0]
            if stat >= p:
                components.extend((i,j))
    if 0 not in components:
        components.append(0)
    if 1 not in components:
        components.append(1)
    return components
```

```python
def filterComponents(components, sigma, p):   
    for i, j in zip(components[0::2], components[1::2]):
        pers = (sigma[i]+sigma[i+1])/sum(sigma)
        if pers < p:
            components.pop(components.index(i))
            components.pop(components.index(j))
    return components
```
