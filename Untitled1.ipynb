```python
def ts(Xi):
    """
    Усредняем побочные диоганали элементарной матрицы 
    и переводим в временной ряд
    """
    Xrev = Xi[::-1]
    return np.array([Xrev.diagonal(i).mean() for i in range(-Xi.shape[0]+1, Xi.shape[1])])

# plt.plot(t,xToTs(X[8]))
```

```python
def ssa(F, L):
    N = len(F) 
    K = N - L + 1 # Количество столбцов в траекторной матрице
    X = np.column_stack([F[i:i+L] for i in range(0,K)])
    d = np.linalg.matrix_rank(X) 
    print("d",d)
    # Сингулярное разложение
    U, Sigma, V = np.linalg.svd(X)
    V = V.T 
    X = np.array([Sigma[i] * np.outer(U[:,i], V[:,i]) for i in range(0, L)])
    return X ,sigma
    # XElem.shape
    
    

def nans(dims):
    """
    nans((M,N,P,...)) is an M-by-N-by-P-by-... array of NaNs.
    :param dims: dimensions tuple
    :return: nans matrix
    """
    return np.nan * np.ones(dims)
def ssa_predict(x, dim, k, n_forecast, e=None, max_iter=10000) -> np.ndarray:
    """
    Series data prediction based on SSA
    Example:
    >>> x = np.linspace(0,5,1000)
    >>> y = 2*x + 2*np.sin(5*x) + 0.5*np.random.randn(1000)
    >>>
    >>> # make prediction for next 200 points
    >>> ys = ssa_predict(y, 100, [0,1,2], 200, 0.01)
    >>> 
    >>> # plot prediction
    >>> plt.figure(figsize=(16,8));
    >>> x0 = len(y); plt.plot(range(x0), y); plt.plot(range(x0, x0 + len(ys)), ys, 'r--')
    >>> plt.legend(['Data', 'Forecast'])
    :param x: series to be predicted
    :param dim: the embedding dimension
    :param k: components indexes for reconstruction
    :param n_forecast: number of points to forecast
    :param e: minimum value to ensure convergence
    :param max_iter: maximum number of iterations
    :return: forecasted series
    """
    if not e:
        e = 0.0001 * (np.max(x) - np.min(x))
    mean_x = x.mean()
#     x = x - mean_x
    xf = nans(n_forecast)

    for i in range(n_forecast):
        # here we use previous value as initial estimation
        x = np.append(x, x[-1])
        yq = x[-1]
        y = yq + 2 * e
        n_iter = max_iter
        while abs(y - yq) > e:
            yq = x[-1]

            xr = ssa(x, dim)
            xr = ts(sum(xr[0:5]))

            y = xr[-1]
            x[-1] = y
            n_iter -= 1
            if n_iter <= 0:
                print('ssa_predict> number of iterations exceeded')
                break

        xf[i] = x[-1]
#     xf = xf + mean_x
    return xf
```

```python
def getComponents(X, p):
    d = X.shape[0]
    components = []
    for i in range(d):
        for j in range(i+1, d):
            stat = scipy.stats.pearsonr(ts(X[i]), ts(X[j]))[0]
            if stat >= p:
                components.extend((i,j))
    if 0 not in components:
        components.append(0)
    if 1 not in components:
        components.append(1)
    return components
```

```python
def filterComponents(components, sigma, p):   
    for i, j in zip(components[0::2], components[1::2]):
        pers = (sigma[i]+sigma[i+1])/sum(sigma)
        if pers < p:
            components.pop(components.index(i))
            components.pop(components.index(j))
    return components
```
