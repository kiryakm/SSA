```python
import numpy as np
import matplotlib.pyplot as plt
import os
import operator
import pandas as pd


def XtoTS(Xi):
    """
    Усредняем побочные диоганали элементарной матрицы 
    и переводим в временной ряд
    """
    Xrev = Xi[::-1]
    return np.array([Xrev.diagonal(i).mean() for i in range(-Xi.shape[0]+1, Xi.shape[1])])

def getRemainder(f, fRestored):
    """
    Посчитать абсолютный и относительный остаток
    средний остаток и макс. остаток
    """
    remainder = list()
    percent = list()
    for i in range(len(f)):
        a = f[i]
        b = fRestored[i]
        remainder.append(abs(a - b))
        if a != 0:
            percent.append(abs((a - b) / a) * 100)
            if percent[len(percent)-1] > 1000000:
                percent[len(percent)-1] = 0
    index, value = max(enumerate(remainder), key=operator.itemgetter(1))
    print("Absolute", value)
    index, value = max(enumerate(percent), key=operator.itemgetter(1))
    print("Relative", value)
    print("Average absolute",sum(remainder)/len(remainder))
    print("Average relative", sum(percent)/len(percent),"\n")
    
def Hankelise(X):
    """
    Hankelises the matrix X, returning H(X).
    """
    L, K = X.shape
    transpose = False
    if L > K:
        # The Hankelisation below only works for matrices where L < K.
        # To Hankelise a L > K matrix, first swap L and K and tranpose X.
        # Set flag for HX to be transposed before returning. 
        X = X.T
        L, K = K, L
        transpose = True

    HX = np.zeros((L,K))
    
    # I know this isn't very efficient...
    for m in range(L):
        for n in range(K):
            s = m+n
            if 0 <= s <= L-1:
                for l in range(0,s+1):
                    HX[m,n] += 1/(s+1)*X[l, s-l]    
            elif L <= s <= K-1:
                for l in range(0,L-1):
                    HX[m,n] += 1/(L-1)*X[l, s-l]
            elif K <= s <= K+L-2:
                for l in range(s-K+1,L):
                    HX[m,n] += 1/(K+L-s-1)*X[l, s-l]
    if transpose:
        return HX.T
    else:
        return HX

    
```

```python
def pirs(X, Y):
    xAv = np.mean(X)
    yAv = np.mean(Y)
    return sum((X - xAv) * (Y - yAv)) /\
    np.sqrt(sum((X - xAv) ** 2) * sum((Y - yAv) ** 2))

def rec(XElem, p, d):
# p = 0.8
#     d = np.linalg.matrix_rank(XElem[0]) # The intrinsic dimensionality of the trajectory space.
    D = np.empty([d,d])
    for i in range(d):
        for j in range(i, d):
            a = pirs(XtoTS(XElem[i]), XtoTS(XElem[j]))
            if a >= p:
                D[i][j] = a
            else:
                D[i][j] = 0
    end = []
    end.append(XtoTS(XElem[0]))
    for i in range(d):
        for j in range(i+1, d):
            if D[i][j]!=0:
                end.append(XtoTS(XElem[i]+XElem[j]))
    return sum(end)

def ssa(F, L):
    K = len(F) - L + 1
    # Преобразуем ряд в траекторную матрицу
    X = np.column_stack([F[i:i+L] for i in range(0,K)])
    d = np.linalg.matrix_rank(X) # The intrinsic dimensionality of the trajectory space.
    # Сингулярное разложение
    U, Sigma, V = np.linalg.svd(X)
    V = V.T 
#     return U, Sigma, V
    # Высчитываем элементарные матрицы Xi
    X = np.array([Sigma[i] * np.outer(U[:,i], V[:,i]) for i in range(0, L)])
    print(X.shape)
    return X, d
```
